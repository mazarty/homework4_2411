```python
#задание1
#Пункт 4. Пользователь по команде "ads" может добавить новую полку
import copy

documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}

#Модифицируем функцию polka под две задачи. Если polk_n заполнено идем по шагу - добавление полки
com  = ''


def polka (doc_n=None, polk_n=None):
    res= ''
    directories2 = copy.copy(directories)
    if polk_n is None and doc_n is not None:
        for dir_n, dir_c in directories.items():
            if doc_n in dir_c:
                res=dir_n
                return res
    #если не был найден элемент, равный doc_num (doc_num не пустое) - возвращаем пусто
        return None
    elif polk_n is not None:
        for dir_n in directories2.keys():
            if dir_n == polk_n:
                return 0
        directories[polk_n]=[]
        return 1
    else: return None

while com != 'q':
    com  = input('Введите команду:')
    if com == 's':         
        doc_num = input('Введите номер документа: ')
        if polka(doc_num) is None: print('Документ не найден в базе')
        else: print('Документ хранится на полке: ', polka(doc_num))
    if com == 'ads':         
        polk_num = input('Введите номер полки: ')
        out = polka(None, polk_num ) 
        if out == 0: print('Такая полка уже существует. Текущий перечень полок:', ', '.join(directories.keys()) )
        if out == 1: print('Полка добавлена. Текущий перечень полок:', ', '.join(directories.keys()) )

#print(directories)
```
